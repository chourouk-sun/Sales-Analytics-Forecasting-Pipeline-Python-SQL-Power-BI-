```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("clean_sales_data.csv")
df['Order Date'] = pd.to_datetime(df['Order Date'])

ts = (
    df.groupby('Order Date')['Sales']
    .sum()
    .asfreq('D')
    .fillna(0)
)

ts.plot(title="Daily Sales")
plt.show()
```

```python
#import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("clean_sales_data.csv")
df['Order Date'] = pd.to_datetime(df['Order Date'])

ts = (
    df.groupby('Order Date')['Sales']
    .sum()
    .asfreq('D')
    .fillna(0)
)

ts.plot(title="Daily Sales")
plt.show()
```

```python
#  Simple Forecast (Moving Average)
ts_ma = ts.rolling(window=30).mean()

ts.plot(label='Actual Sales')
ts_ma.plot(label='30-Day Moving Average')
plt.legend()
plt.title("Sales Forecast (Moving Average)")
```

```python
#Step 1: Load clean data
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("clean_sales_data.csv")
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python

# Create Time Series
ts = (
    df.groupby('Order Date')['Sales']
    .sum()
    .asfreq('D')
    .fillna(0)
)
```

```python

# Plot & Basic Forecast
ts.plot(title="Daily Sales Time Series")
plt.show()

ts.rolling(30).mean().plot(label="30-Day Moving Average")
plt.legend()
plt.show()
```
